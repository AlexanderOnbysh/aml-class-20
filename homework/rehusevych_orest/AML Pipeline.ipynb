```python
# Task 2
import yaml

def read_mappings(path_to_yaml):
    
    with open(path_to_yaml, 'r') as file:
        features_mapping = yaml.safe_load(file)

    target_feature = "loss"
    numeric_features = [column for column in features_mapping.keys() if features_mapping[column] == "Numeric"]
    categorical_features = [column for column in features_mapping.keys() if features_mapping[column] == "Categorical"]
    text_features = [column for column in features_mapping.keys() if features_mapping[column] == "Text"]

    if target_feature in numeric_features:      numeric_features.remove(target_feature);
    if target_feature in categorical_features:  categorical_features.remove(target_feature);
    if target_feature in text_features:         text_features.remove(target_feature);
        
    return numeric_features, categorical_features, text_features
```

```python
# Task 4

from sklearn.model_selection import train_test_split
from copy import deepcopy


def cv(demo_fire_df, target_feature):

    y = demo_fire_df.pop(target_feature)
    X = deepcopy(demo_fire_df)

    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

    # y should be an ndarray
    y_train = y_train.values
    y_test = y_test.values
    
    return X_train, X_test, y_train, y_test
```

```python
from sklearn.base import BaseEstimator, TransformerMixin, clone
from sklearn.linear_model import ElasticNet
from sklearn.model_selection import KFold


class StackingTransformer(BaseEstimator, TransformerMixin):
    """A transformer that wraps a RegressorMixin `est` predict function.
    
    It overloads `fit_transform` to do a stacked transform to avoid leakage.
    """
    def __init__(self, est):
        self.est = est
        
    def fit(self, X, y=None):
        if y is None:
            raise ValueError('{}.fit requires y to be not None'.format(self))
        self.est.fit(X, y)
        return self
    
    def transform(self, X):
        return self.est.predict(X)[:, np.newaxis]
    
    def fit_transform(self, X, y=None):
        if y is None:
            raise ValueError('{}.fit requires y to be not None'.format(self))
        self.models_ = []
        out = np.empty_like(y)[:, np.newaxis]
        for train, test in KFold(5, shuffle=True, random_state=0).split(X):
            self.models_.append(clone(self.est).fit(X[train], y[train]))
            out[test, 0] = self.models_[-1].predict(X[test])
        self.est = self.models_[0]
        return out
```

```python
# Task 5

from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.compose import ColumnTransformer
from sklearn.feature_selection import SelectFromModel

from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor


def create_pipeline(model_tag):

    # Numeric pipeline
    numeric_pipeline = Pipeline(
        steps=[
            ('num_imputer', SimpleImputer())
        ]
    )


    # Categorical pipeline
    if model_tag == 'lr':
        categorical_pipeline = Pipeline(
            steps=[
                ('cat_imputer', SimpleImputer(strategy='constant', fill_value='Missing')),
                ('cat_one_hot', OneHotEncoder(handle_unknown='ignore')),
                ('cat_select',  SelectFromModel(ElasticNet(alpha=0.1), threshold=None))
            ]
        )
    elif model_tag == 'rf':
        categorical_pipeline = Pipeline(
            steps=[
                ('cat_imputer', SimpleImputer(strategy='constant', fill_value='Missing')),
                ('cat_one_hot', OneHotEncoder(handle_unknown='ignore')),
                ('cat_select',  SelectFromModel(RandomForestRegressor(), threshold=None))
            ]
        )
    elif model_tag == 'gb':
        categorical_pipeline = Pipeline(
            steps=[
                ('cat_imputer', SimpleImputer(strategy='constant', fill_value='Missing')),
                ('cat_one_hot', OneHotEncoder(handle_unknown='ignore')),
                ('cat_select',  SelectFromModel(GradientBoostingRegressor(), threshold=None))
            ]
        )
    else:
        raise ValueError("model tag should be either lr, rf or gb")


    # Text pipeline
    text_pipeline = Pipeline(
        steps=[
            ('text_bow',     CountVectorizer()),
            ('text_bow_reg', StackingTransformer(ElasticNet(alpha=0.1)))
        ]
    )


    # Preprocessor
    preprocessor = ColumnTransformer(
        transformers=[
            ('numeric',     numeric_pipeline,     numeric_features),
            ('categorical', categorical_pipeline, categorical_features),
            ('text',        text_pipeline,        text_features[0])
        ]
    )


    # Creating model
    if model_tag == 'lr':
        model = LinearRegression
    elif model_tag == 'rf':
        model = RandomForestRegressor
    elif model_tag == 'gb':
        model = GradientBoostingRegressor
    else:
        raise ValueError("model tag should be either lr, rf or gb")

    # Pipeline
    if model_tag == 'gb':
        pipeline = Pipeline(
            steps=[
                ('preprocessor', preprocessor),
                (model_tag, model())
            ]
        )
    else:
        pipeline = Pipeline(
            steps=[
                ('preprocessor', preprocessor),
                (model_tag, model(n_jobs=-1))
            ]
        )
    
    return pipeline
```

```python
# Task 6

from skopt import BayesSearchCV
from sklearn.model_selection import RandomizedSearchCV



def optimize(pipeline, model_tag, X_train, y_train):
    
    # Parameter grid for TF-IDF
    param_grid = {
        'preprocessor__text__text_bow__ngram_range': [(1, 1), (1, 2), (2, 2), (1, 3), (2, 3)],
        'preprocessor__text__text_bow__analyzer': ['word', 'char', 'char_wb']
    }

    # Parameter grid for Random Forest
    if model_tag == 'rf':
        param_grid.update(
            {
                'rf__n_estimators': range(5, 100), 
                'rf__max_features': range(1, 20),
                'rf__max_depth': range(2, 10)
            }
        )
    elif model_tag == 'gb':
        param_grid.update(
            {
                'gb__n_estimators': range(5, 100), 
                'gb__max_features': ['auto', 'sqrt', 'log2'],
                'gb__max_depth': range(2, 10),
                'gb__learning_rate': np.linspace(0.00001, 0.1, 4),
                'gb__criterion': ['friedman_mse', 'mse', 'mae']
            }
        )

    # Random Search
    random_search = RandomizedSearchCV(pipeline, param_distributions=param_grid, cv=5, scoring='neg_mean_absolute_error')
    random_search.fit(X_train, y_train);
    return random_search

    # Bayesian Optimization
    # bayes_search = BayesSearchCV(pipeline, search_spaces=param_grid, scoring='neg_mean_absolute_error')
    # bayes_search.fit(X_train, y_train)
```

```python
# Task 7

from sklearn.metrics import mean_absolute_error

def show_results(model_tag, random_search, X_test, y_test):
    print(f'Model: {model_tag}')
    print("MAE: %.3f" % mean_absolute_error(y_test, random_search.best_estimator_.predict(X_test)))
    print("Best params: {}".format(random_search.best_params_))
```

```python
def full_pipeline(model_tag):
    link_to_data = 'https://s3.amazonaws.com/datarobot_public_datasets/DR_Demo_Fire_Ins_Loss_only.csv'
    path_to_mappings = "features_mapping.yaml"
    target_feature = 'loss'
    
    # Task 1
    df = pd.read_csv(link_to_data)
    
    # Task 2
    num_f, cat_f, text_f = read_mappings(path_to_mappings)
    
    # Task 3 (Optional)
    
    # Task 4
    X_train, X_test, y_train, y_test = cv(df, target_feature)
    
    # Task 5
    pipeline = create_pipeline(model_tag)
    
    # Task 6
    random_search = optimize(pipeline, model_tag, X_train, y_train)
    
    # Task 7
    show_results(model_tag, random_search, X_test, y_test)
```
